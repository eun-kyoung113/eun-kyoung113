<p>자연수 n을 뒤집어 각 자리 숫자를 원소로 가지는 배열 형태로 리턴해주세요. 예를들어 n이 12345이면 [5,4,3,2,1]을 리턴합니다.</p>

<h5>제한 조건</h5>

<ul>
<li>n은 10,000,000,000이하인 자연수입니다.</li>
</ul>

<h5>입출력 예</h5>

<table class="table">
<thead><tr>
<th>n</th>
<th>return</th>
</tr>
</thead>
<tbody><tr>
<td>12345</td>
<td>[5,4,3,2,1]</td>
</tr>
</tbody>
</table>

```python
def solution(n:int):
    value=str(n)
    result=[]
    for i in range((len(value)-1),-1,-1):
        result.append(int(value[i]))
    return(result)
```

```python
solution(14567)
```
