<p>정수 n을 입력받아 n의 약수를 모두 더한 값을 리턴하는 함수, solution을 완성해주세요.</p>

<h5>제한 사항</h5>

<ul>
<li><code>n</code>은 0 이상 3000이하인 정수입니다.</li>
</ul>

<h6>입출력 예 설명</h6>

<p>입출력 예 #1<br/>
12의 약수는 1, 2, 3, 4, 6, 12입니다. 이를 모두 더하면 28입니다.</p>

<p>입출력 예 #2<br/>
5의 약수는 1, 5입니다. 이를 모두 더하면 6입니다.</p>



```python
def solution(num:int):
    root_num=int(num**0.5)
    answer=0
    for i in range(1,(root_num+1)):
        if (num%i)==0:
            value=int(num/i)
            answer+=i
            answer+=value
            if value==i: # 25=5X5의 경우, value와 i 값이 동일하게 된다.
                answer-=i
        
    return answer
```

```python
solution(10)
```
