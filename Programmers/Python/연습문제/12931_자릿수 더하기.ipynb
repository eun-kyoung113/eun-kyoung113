<p>자연수 N이 주어지면, N의 각 자릿수의 합을 구해서 return 하는 solution 함수를 만들어 주세요.<br/>
예를들어 N = 123이면 1 + 2 + 3 = 6을 return 하면 됩니다.</p>

<h5>제한사항</h5>

<ul>
<li>N의 범위 : 100,000,000 이하의 자연수</li>
</ul>

<hr/>

<h5>입출력 예 설명</h5>

<p>입출력 예 #1<br/>
문제의 예시와 같습니다.</p>

<p>입출력 예 #2<br/>
9 + 8 + 7 = 24이므로 24를 return 하면 됩니다.</p>



```python
def solution(N:int):
    n=str(N)
    result=0
    for value in n:
        result+=int(value)
    return result
```

```python
solution(101)
```
