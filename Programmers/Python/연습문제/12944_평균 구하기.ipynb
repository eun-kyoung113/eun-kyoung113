<p>정수를 담고 있는 배열 arr의 평균값을 return하는 함수, solution을 완성해보세요.</p>

<h4>제한사항</h4>

<ul>
<li>arr은 길이 1 이상, 100 이하인 배열입니다.</li>
<li>arr의 원소는  -10,000 이상 10,000 이하인 정수입니다.</li>
</ul>

<h4>입출력 예</h4>

<table class="table">
<thead><tr>
<th>arr</th>

</tr>
</thead>

```python
def solution(arr):
    return(sum(arr)/len(arr))
```

```python
solution([1,2,3,5])
```
