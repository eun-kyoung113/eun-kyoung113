<p>단어 s의 가운데 글자를 반환하는 함수, solution을 만들어 보세요. 단어의 길이가 짝수라면 가운데 두글자를 반환하면 됩니다.</p>

<h6>재한사항</h6>

<ul>
<li>s는 길이가 1 이상, 100이하인 스트링입니다.</li>
</ul>

<h5>입출력 예</h5>

<table class="table">
<thead><tr>
<th>s</th>
<th>return</th>
</tr>
</thead>
<tbody><tr>
<td>"abcde"</td>
<td>"c"</td>
</tr>
<tr>
<td>"qwer"</td>
<td>"we"</td>
</tr>
</tbody>
</table>

```python
def solution(s:str):
    if len(s)%2==0:
        result=s[int((len(s)/2)-1):int(len(s)/2)+1]
    else:
        ind=int(len(s)/2)
        result=s[ind]
    return result
```

```python
solution('abcdef')
```
