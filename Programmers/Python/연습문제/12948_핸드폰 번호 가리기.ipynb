<p>프로그래머스 모바일은 개인정보 보호를 위해 고지서를 보낼 때 고객들의 전화번호의 일부를 가립니다.<br/>
전화번호가 문자열 phone_number로 주어졌을 때, 전화번호의 뒷 4자리를 제외한 나머지 숫자를 전부 <code>*</code>으로 가린 문자열을 리턴하는 함수, solution을 완성해주세요.</p>

<h5>제한 조건</h5>

<ul>
<li>phone_number는 길이 4 이상,  20이하인 문자열입니다.</li>
</ul>

<h5>입출력 예</h5>

<table class="table">
<thead><tr>
<th>phone_number</th>
<th>return</th>
</tr>
</thead>
<tbody><tr>
<td>"01033334444"</td>
<td>"*******4444"</td>
</tr>
<tr>
<td>"027778888"</td>
<td>"*****8888"</td>
</tr>
</tbody>
</table>

```python
def solution(phone_number:str):
    result="*"*(len(phone_number)-4)+phone_number[len(phone_number)-4:]
    return result
```

```python
solution("01023456790")
```
