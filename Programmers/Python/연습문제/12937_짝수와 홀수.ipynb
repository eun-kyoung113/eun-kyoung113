<p>정수 num이 짝수일 경우 "Even"을 반환하고 홀수인 경우 "Odd"를 반환하는 함수, solution을 완성해주세요.</p>

<h5>제한 조건</h5>

<ul>
<li>num은 int 범위의 정수입니다.</li>
<li>0은 짝수입니다.</li>
</ul>

<h5>입출력 예</h5>

```python
def solution(num:int):
    if (num%2==0):
        return "Even"
    else:
        return "Odd"
```

```python
solution(1356)
```
